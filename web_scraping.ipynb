```python
from bs4 import BeautifulSoup
import requests
```

```python
# Define the URL and make an HTTP request to fetch the HTML content of the page

url = 'https://en.wikipedia.org/wiki/List_of_most-visited_art_museums'

page = requests.get(url)

# Create a Beautiful Soup object to facilitate navigation

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
# Find all tables, and the index shows the first one found

soup.find_all('table')[0]
```

```python
# Find a specific table
soup.find('table', class_ = 'wikitable sortable')
```

```python
# Store the desired table in a variable

table = soup.find_all('table')[0]
```

```python
print(table)
```

```python
# Display all the column titles of the table we defined. If we used soup.find_all, it would show all the titles in the document.

titles = table.find_all('th')
```

```python
# Clean and organize the extracted titles into a list

table_titles = [title.text.strip() for title in titles]

print(table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = table_titles)

df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  print(individual_row_data)

  # Note that the first item that appears is the empty bracket, making the next step (locating and adding to our DataFrame) impossible since the titles and the information wouldn't match in quantity.
```

```python
df
```

```python
# The index will resolve this issue by accessing from the second item onwards.
for row in column_data[1:]:
  row_data = row.find_all('td')
  individual_row_data = [data.text.strip() for data in row_data]
  print(individual_row_data)

  lenght = len(df)
  df.loc[lenght] = individual_row_data
```

```python
df
```

Here we have another small issue: the "Visitors annually" column does not present only the number of visitors, making it difficult to filter by visitor count, for example. To address this, let's first verify the data type we are dealing with and then create a function to clean each row in this column.

```python
print(df.dtypes)
```

```python
def clean_visitors_column(column):
  # Converting to text
   column = column.astype(str)
   # Extracting only the numbers
   extracted = column.str.extract(r'(\d[\d,]*)')[0] # This expression in parentheses specifies finding digits from 0 to 9, allowing special characters, commas, and zeros
  # Replacing commas and converting to float
   return extracted.str.replace(',', '').astype(float)

# Here I used .loc to avoid the warning about trying to modify a copy of the DataFrame
# Applying the function to the desired column
df.loc[:, 'Visitors annually'] = clean_visitors_column(df['Visitors annually'])
```

```python
df
```

```python
df_filtered_visitors = df[df['Visitors annually'] > 3000000]

styled_df = df_filtered_visitors.style.set_table_attributes('style="display:inline; margin: 20px;"') \
                               .set_caption('More than 3 Million visitors annually')

styled_df
```
